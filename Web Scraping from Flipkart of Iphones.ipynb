```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://www.flipkart.com/search?q=iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
r=requests.get(url)
print(r)
```

```python
soup=bs(r.text,'html.parser')
print(soup)
```

```python
# Fetching data for one Product
```

```python
# Fetch Iphone Name
```

```python
product = soup.find('div',attrs={ 'class':"_4rR01T"})
print(product.text)
```

```python
# Fetch iphone desc
```

```python
desc = soup.find('div',attrs= {'class':"fMghEO"})
print(desc.text)
```

```python
# Fetch Iphone Rating (star)
```

```python
Rating=soup.find('div',attrs= {'class':"_3LWZlK"})
print(Rating.text)
```

```python
# Fetch Iphone Price
```

```python
Price=soup.find('div' ,attrs= {'class':"_30jeq3 _1_WHN1"})
print(Price.text)
```

```python
link=soup.find('a',attrs={'class':"_1fQZEK"})
print(link)
print("-"*10)
print(link.get('herf'))
```

```python
print(link.text)
```

```python
link=soup.find_all('a',attrs={'class':"_1fQZEK"})
print(link)
print(link.get('herf'))
```

```python
# Fetch all Iphone in the result
```

```python
product = []
desc = []
Rating= []
Price = []
```

```python
for i in soup.findAll('div',attrs={ 'class':"_4rR01T"}):
    product.append(i.text)
print(len(product))
```

```python
for i in soup.findAll('div',attrs= {'class':"fMghEO"}):
    desc.append(i.text)
print(len(desc))
```

```python
for i in soup.findAll('div',attrs= {'class':"_3LWZlK"}):
    Rating.append(i.text)
print(len(Rating))
```

```python
for i in soup.findAll('div' ,attrs= {'class':"_30jeq3 _1_WHN1"}):
    Price.append(i.text)
print(len(Price))
```

```python
df=pd.DataFrame({'Product':product[:20],'Description':desc[:20],'Price':Price[:20],'Rating':Rating[:20]})
df.head(20)
```

```python
df.to_csv('Flipkart_Iphone .csv')
```
