```python
import pandas as pd
import numpy as np
```

```python
# dataFrame 1
d1 = {'Customer_id':[1,2,3,4,5,6],
     'Product':['Oven','Bottle','Bag','Laptop','Television','Mobile']}
df1 = pd.DataFrame(d1)
 
# data frame 2
d2 = {'Customer_id':[2,4,6,7,8],
    'State':['California','Florida','Texas','Austria','Spain']}
df2 = pd.DataFrame(d2)
```

```python
df1
```

```python
df2
```

### Merge

<b>inner : </b> Return only the rows in which the left table has matching keys in the right table

```python
# Inner  
d1 = pd.merge(df1, df2, on='Customer_id', how='inner')
# on = ['Customer_id','Transcation_No']
d1
```

<b>outer :</b> Returns all rows from both tables, join records from the left which have matching keys in the right table.

```python
d2 = pd.merge(df1, df2, on='Customer_id', how='outer')
d2
```

<b>Left outer : </b> Return all rows from the left table, and any rows with matching keys from the right table.

```python
d3 = pd.merge(df1, df2, on='Customer_id', how='left')
d3
```

<b>Right outer :</b> Return all rows from the right table, and any rows with matching keys from the left table.

```python
d4 = pd.merge(df1, df2, on='Customer_id', how='right')
d4
```
