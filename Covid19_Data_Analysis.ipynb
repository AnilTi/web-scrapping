##### Find Out The Discharged Patients From The States Maharashtra,Karnataka,MP,Gujrat.

```python
print('Discharged Patients Of Maharashtra',data['Discharged'][20])
print('Discharged Patirents Of Karnataka',data['Discharged'][15])
print('Discharged Patirents Of MP',data['Discharged'][19])
print('Discharged Patirents Of Gujrat',data['Discharged'][10])
```

#####  Find Out The States Having Total Cases between 1lac to 5lac.

```python
data[data['Total Cases'].between(100000,500000)]
```

#####  Display All The Data Of The States Which Has Active Cases Less Than 1000.

```python
data[data['Active']<1000]
```

##### Make A Separate CSV File Of  Data Of The States Which Has Active Cases Less Than 1000
